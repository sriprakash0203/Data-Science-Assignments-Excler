# Lets assume: taxable_income <= 30000 as â€œRisky=0â€ and others are â€œGood=1â€

```python
print('minimum_value : ' , df['Taxable.Income'].min() ,'\n maximun_value :',df['Taxable.Income'].max())
```

```python
#Converting Target variable 'Sales' into categories Low, Medium and High.
df['Taxable.Income'] = pd.cut(x=df['Taxable.Income'],bins = [10002,30000,99620], labels=['Risky','Good'])
df['Taxable.Income']
```

```python
df.head()
```

```python
sns.countplot(df['Taxable.Income'])
plt.grid(True)
plt.show()
```

```python
df['Taxable.Income'].value_counts()
```

```python
sns.countplot(df['Taxable.Income_Good'])
plt.grid(True)
plt.show()
```

```python
#encoding categorical fraud_data
label_encoder = preprocessing.LabelEncoder()

df['Undergrad'] = label_encoder.fit_transform(df['Undergrad'])
df['Taxable.Income_Good'] = label_encoder.fit_transform(df['Taxable.Income_Good'])
df['Marital.Status'] = label_encoder.fit_transform(df['Marital.Status'])
df['Urban'] = label_encoder.fit_transform(df['Urban'])
```

```python
df
```

```python
df['Taxable.Income_Good'].value_counts()
```

```python
X=df.drop('Taxable.Income_Good',axis=1)
y=df[['Taxable.Income_Good']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33, random_state = 12)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

# Building Decision Tree Classifier using Entropy Criteria

```python
dt_model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
dt_model.fit(X_train,y_train)
```

```python
DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
plt.figure(figsize=(16,8))
tree.plot_tree(dt_model)
plt.show()
```

```python
y_pred_train = dt_model.predict(X_train)
```

```python
y_pred_test = dt_model.predict(X_test)
y_pred_test
```

Train Data

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
confusion_matrix(y_train,y_pred_train)
```

```python
accuracy_score(y_test,y_pred_test)
```

```python
confusion_matrix(y_test,y_pred_test)
```
